```python
import sympy  as sp
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Gráfica de contagios diarios
```

```python
k=0.117
c=19999
t=np.arange(0,280)
C=20000/(1+c*np.exp(-k*t))
```

```python
dC_dt=k*C*(1-(C/20000))
```

```python
plt.title("Comportamiento de contagios")
plt.plot(C,dC_dt, color="Green")
plt.legend=("Max de contagidos")
plt.axhline(585,color="blue")
plt.axvline(10000,color="yellow")
plt.xlabel("Poblacion propensa al contagio")
plt.ylabel("dC/dt")
```

```python
plt.plot(t,C)
plt.ylabel("N° de contagiados")
plt.xlabel("N° de días")
```
